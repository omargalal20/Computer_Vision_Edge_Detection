```python
#Function 1: Given an image, returns the image and it's squared version in array format
def toArrayAndSquare(im):
  img = [[0 for x in range(im.size[0])] for y in range(im.size[1])]
  sqr = [[0 for x in range(im.size[0])] for y in range(im.size[1])]
  for i in range (0,im.size[0]):
    for j in range (0,im.size[1]):
      img[j][i] = im.getpixel((i,j))
      sqr[j][i] = img[j][i] ** 2
  return img,sqr

#Function 2: Given an image, applies a certain threshold
def applyThreshold (im, th):
  res = [[0 for x in range(len(im[0]))] for y in range(len(im))]
  for i in range (0,len(im)):
    for j in range (0,len(im[0])):
      if(im[i][j]<th):
        res[i][j] = 0
      else:
        res[i][j] = 255
  return res
```

```python
#Function 3: Given an image, returns its unique values
def getUniqueValues(image):
    imgDf = pd.DataFrame(np.array(image))
    a = np.array([])
    for i in range(imgDf.shape[0]):
        a = np.hstack((a, np.array(imgDf.iloc[i].unique())))
    return np.unique(a)

#Function 4: Given an image, returns its empty co occurrence matrix
def createEmptyCoOccurrenceMatrix(image):
    uniqueValuesOfImage = getUniqueValues(image)
    mat = np.zeros(uniqueValuesOfImage.shape[0], dtype=int)
    for _ in range(1, uniqueValuesOfImage.shape[0]):
        mat = np.vstack((mat, np.zeros(uniqueValuesOfImage.shape[0], dtype=int)))
    return pd.DataFrame(mat)
```

```python
#Function 5: Given an image, returns its co occurence matrix
def calculateCooccurrence(image,  relationship):
    coOccurrenceMat = createEmptyCoOccurrenceMatrix(img)
    if(relationship == 'Horizontal'):
        print(relationship)
        for i in range (0, len(image) - 1):
            for j in range (0, len(image[0])):
                coOccurrenceMat[image[i][j]][image[i+1][j]] = coOccurrenceMat[image[i][j]][image[i+1][j]] + 1
    else:
        print(relationship)
        for i in range (0, len(image)):
            for j in range (0, len(image[0]) - 1):
                coOccurrenceMat[image[i][j]][image[i][j + 1]] = coOccurrenceMat[image[i][j]][image[i][j + 1]] + 1
    return coOccurrenceMat
```

```python
#Function 5: Given an Co-occurrence Matrix of an Image, returns the Co-occurrence Matrix with the pixels of some radius/window-size on its diagonal set to 0
def nullifyPixels(coOccurrence):
    coOccurrence_mat = coOccurrence.copy()
    for i in range (0, coOccurrence_mat.shape[0]):
        for j in range (0, coOccurrence_mat.shape[1]):
            if(abs(i - j) < 30):
                coOccurrence_mat[i][j] = 0
    return coOccurrence_mat

horizontalNullified = nullifyPixels(horizontal)
verticalNullified = nullifyPixels(vertical)
```

```python
#Function 5: Given an Image and it's Co-occurrence Matrices, returns the reconstructed image
def imgWithCooccurrence(image, horizontalNullified, verticalNullified):
    reconstructedImage = [[0 for _ in range(len(img[0]))] for _ in range(len(img))]
    for i in range (0, len(image) - 1):
        for j in range (0, len(image[0]) - 1):
            if(horizontalNullified[image[i][j]][image[i+1][j]] > 0 or verticalNullified[image[i][j]][image[i][j + 1]] > 0):
                reconstructedImage[i][j] = 255
            else:
                reconstructedImage[i][j] = 0
    return reconstructedImage

reconstructedImage = imgWithCooccurrence(img, horizontalNullified, verticalNullified)
```

```python
def s(image_matrix):
  s_mat = [[0 for x in range(len(image_matrix[0]))] for y in range(len(image_matrix))]
  for i in range (0, len(image_matrix)):
    for j in range (0, len(image_matrix[0])):
      if(j == 0):
        s_mat[i][j] = image_matrix[i][j]
      else:
        s_mat[i][j] = s_mat[i][j-1] + image_matrix[i][j]
  return s_mat
```

```python
def integralArray(img_matrix):
  ii_mat = [[0 for x in range(len(img_matrix[0]))] for y in range(len(img_matrix))]
  s_mat = s(img_matrix)
  for i in range (0,len(img_matrix)):
      for j in range (0,len(img_matrix[0])):
          if(i == 0):
            ii_mat[i][j] = s_mat[i][j]
          else:
            ii_mat[i][j] = ii_mat[i-1][j] + s_mat[i][j]
  return ii_mat
```

```python
def localSum(integral_img , topLeft , bottomRight):
  removeAbove = False
  removeLeft = False
  sum = integral_img[bottomRight[0]][bottomRight[1]] 
  if( topLeft[1] > 0 ):
    sum = sum - integral_img[bottomRight[0]][topLeft[1]-1]
    removeLeft = True
  if( topLeft[0] > 0 ):
    sum = sum - integral_img[topLeft[0]-1][bottomRight[1]]
    removeAbove = True
  if (removeAbove and removeLeft):
    sum = sum + integral_img[topLeft[0]-1][topLeft[1]-1]
  return sum
```

```python
img,sqr = toArrayAndSquare(grayScaleImage)
```

```python
img_integral = integralArray(img)
sqr_integral = integralArray(sqr)
```

```python
def imgWithIntegral(img,rows , cols):
  img_integral = integralArray(img)
  sqr_integral = integralArray(sqr)
  output_img = img
  start_row_index = rows//2 
  start_col_index = cols//2 
  for i in range(start_row_index,len(img)-start_row_index):
    for j in range(start_col_index,len(img[0]) - start_col_index):
      sum_img = localSum(img_integral , [i-rows//2,j-cols//2] , [i+rows//2,j+cols//2])
      sum_sqr = localSum(sqr_integral ,[i-rows//2,j-cols//2] , [i+rows//2,j+cols//2])
      variance = (sum_sqr/(rows*cols)) - ((sum_img/(rows*cols))**2)
      if(variance<0):
        variance = 0
      output_img[i][j] = variance
  return output_img
```

```python
img_variance = imgWithIntegral(img,3,3)
```
